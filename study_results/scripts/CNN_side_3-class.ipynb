```python
# training script for side 3-class classification CNN, without 5-fold CV, with HOG features, no hyperparameters selected
# several metrics saved and plotted

import os
import numpy as np
import cv2
import pandas as pd
from sklearn.metrics import ConfusionMatrixDisplay, accuracy_score, confusion_matrix, precision_recall_fscore_support
from sklearn.model_selection import train_test_split
from sklearn.utils import class_weight
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
import matplotlib.pyplot as plt
from skimage import exposure
from skimage.feature import hog
from tqdm import tqdm
import concurrent.futures
import json

# Paths to the folders containing the images and height arrays
image_folder = 'bright_spicy_2024/CNN_side_images'
height_folder = 'bright_spicy_2024/CNN_height_arrays'

# Paths to save the model and results
model_save_path = 'CNN_3chp_side_model.pth'
results_save_path = 'CNN_results/3c_side_CNN_results.json'
metrics_save_path = 'CNN_results/3c_side_CNN_metrics.json'
bar_plot_save_path = 'CNN_results/3c_side_CNN_actual_vs_predicted_bar_plot.png'
randomized_metrics_save_path = 'CNN_results/3c_side_CNN_randomized_metrics.json'
randomized_bar_plot_save_path = 'CNN_results/3c_side_CNN_randomized_actual_vs_predicted_bar_plot.png'

# Function to calculate the trimmed mean of a height array
def calculate_trimmed_mean(height_array, trim_bottom=0.08805012610937063, trim_top=0.2893423891188367):
    sorted_height_array = np.sort(height_array)
    lower_bound = int(trim_bottom * len(sorted_height_array))
    upper_bound = int((1 - trim_top) * len(sorted_height_array))
    
    if trim_bottom + trim_top >= 1.0:
        raise ValueError("The sum of trim_bottom and trim_top must be less than 1.0")
    
    trimmed_array = sorted_height_array[lower_bound:upper_bound]
    trimmed_mean = np.mean(trimmed_array)
    
    return trimmed_mean

# Helper functions
def load_image_and_height(height_file, image_folder, height_folder):
    height_path = os.path.join(height_folder, height_file)
    side_image_name = os.path.splitext(height_file)[0]
    img_path = os.path.join(image_folder, side_image_name + '.jpg')

    if not os.path.exists(img_path):
        print(f"Image file does not exist: {img_path}")
        return None, None, None

    img = cv2.imread(img_path)
    if img is None:
        print(f"Failed to load image: {img_path}")
        return None, None, None
    
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    height, width = img.shape[:2]
    start_x = (width - 3280) // 2
    start_y = (height - 2464) // 2
    img = img[start_y:start_y + 2464, start_x:start_x + 3280]
    img = cv2.resize(img, (224, 224))

    height_array = np.load(height_path)
    if height_array.size == 0:
        print(f"Height array is empty: {height_path}")
        return None, None, None

    mean_height = calculate_trimmed_mean(height_array)
    hog_features, _ = extract_hog_features(img)
    return img, mean_height, hog_features

def extract_hog_features(img):
    hog_features, hog_image = hog(img, orientations=9, pixels_per_cell=(8, 8), cells_per_block=(2, 2), visualize=True, channel_axis=-1)
    return hog_features, hog_image

def preprocess_data(height_files, image_folder, height_folder):
    images, mean_heights, hog_features_list = [], [], []
    with concurrent.futures.ThreadPoolExecutor() as executor:
        results = list(executor.map(lambda f: load_image_and_height(f, image_folder, height_folder), height_files))
    for img, mean_height, hog_features in results:
        if img is not None and mean_height is not None:
            images.append(img)
            mean_heights.append(mean_height)
            hog_features_list.append(hog_features)
    return np.array(images), np.array(mean_heights), np.array(hog_features_list)

# List all height files and match with side images
height_files = [f for f in os.listdir(height_folder) if f.endswith('.npy')]
side_images = [os.path.splitext(f)[0] for f in height_files]

# Check if all side images are available
side_image_files = [f for f in os.listdir(image_folder) if f.endswith('.jpg')]
missing_images = set(side_images) - set(os.path.splitext(f)[0] for f in side_image_files)

if missing_images:
    print(f"Missing images for: {missing_images}")

# Preprocess data
images, mean_heights, hog_features_list = preprocess_data(height_files, image_folder, height_folder)
print(f"Loaded {len(images)} images and {len(mean_heights)} height arrays.")

# Check if we have any data
if len(images) == 0 or len(mean_heights) == 0:
    raise ValueError("No images or height arrays were loaded. Please check the data folder and files.")

# Normalize images
images = images / 255.0

# Adjusted thresholds for better balance
adjusted_short_threshold = 748
adjusted_tall_threshold = 1205

# Label each height array using the adjusted thresholds
def label_height_array(mean_height, short_threshold, tall_threshold):
    if (mean_height < short_threshold):
        return 'Short'
    elif (mean_height > tall_threshold):
        return 'Tall'
    else:
        return 'Medium'

labels = np.array([label_height_array(mean_height, adjusted_short_threshold, adjusted_tall_threshold) for mean_height in mean_heights])
label_map = {'Short': 0, 'Medium': 1, 'Tall': 2}
numerical_labels = np.array([label_map[label] for label in labels])

# Check class balance
unique, counts = np.unique(numerical_labels, return_counts=True)
print("Class distribution:", dict(zip(unique, counts)))

# Visualize resized images with their labels to check correctness
def visualize_samples(images, labels, label_map, num_samples=10):
    plt.figure(figsize=(20, 10))
    for i in range(num_samples):
        idx = np.random.randint(0, len(images))
        img = images[idx]
        label = labels[idx]
        
        plt.subplot(2, num_samples, i+1)
        plt.imshow(img)
        plt.title(f"Resized\nLabel: {list(label_map.keys())[list(label_map.values()).index(label)]}")
        plt.axis('off')
    plt.show()

visualize_samples(images, numerical_labels, label_map)

# Custom preprocessing function for transforms
def custom_preprocessing(img):
    img = np.array(img)  # Convert PIL Image to NumPy array
    for channel in range(img.shape[2]):
        img[..., channel] = exposure.rescale_intensity(img[..., channel], out_range=(0, 1))
        img[..., channel] = exposure.equalize_hist(img[..., channel])
        img[..., channel] = exposure.equalize_adapthist(img[..., channel], clip_limit=0.03)
    return img

# PyTorch Dataset class
class HeightDataset(Dataset):
    def __init__(self, images, hog_features, labels, transform=None):
        self.images = images
        self.hog_features = hog_features
        self.labels = labels
        self.transform = transform

    def __len__(self):
        return len(self.images)

    def __getitem__(self, idx):
        img = self.images[idx]
        hog_feature = self.hog_features[idx]
        label = self.labels[idx]
        if self.transform:
            img = self.transform(img)
        return img, hog_feature, label

# Transforms with additional augmentations
train_transform = transforms.Compose([
    transforms.ToPILImage(),
    transforms.RandomHorizontalFlip(),
    transforms.Lambda(custom_preprocessing),
    transforms.ToTensor()
])

val_test_transform = transforms.Compose([
    transforms.ToPILImage(),
    transforms.Lambda(custom_preprocessing),
    transforms.ToTensor()
])

# Split data for test set
X_train_val, X_test, y_train_val, y_test, hog_train_val, hog_test = train_test_split(images, numerical_labels, hog_features_list, test_size=0.2, stratify=numerical_labels, random_state=42)

# Prepare the training and validation data
X_train, X_val, y_train, y_val, hog_train, hog_val = train_test_split(X_train_val, y_train_val, hog_train_val, test_size=0.2, stratify=y_train_val, random_state=42)

# Check validation data loading
print(f"Validation data shape: {X_val.shape}")
print(f"Validation labels shape: {y_val.shape}")

# Check some validation labels
print("Validation labels sample:", y_val[:10])

# Visualize validation data
visualize_samples(X_val, y_val, label_map)

# Calculate class weights
class_weights = class_weight.compute_class_weight(class_weight='balanced', classes=np.unique(y_train), y=y_train)
class_weights_tensor = torch.tensor(class_weights, dtype=torch.float)

# Datasets and DataLoaders
train_dataset = HeightDataset(X_train, hog_train, y_train, transform=train_transform)
val_dataset = HeightDataset(X_val, hog_val, y_val, transform=val_test_transform)
test_dataset = HeightDataset(X_test, hog_test, y_test, transform=val_test_transform)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True, num_workers=4)
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False, num_workers=4)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False, num_workers=4)

# Define the CNN model
class CNNModel(nn.Module):
    def __init__(self, dropout_rate=0.3, num_layers=3, num_classes=3, hog_feature_size=324):
        super(CNNModel, self).__init__()
        self.cnn_layers = nn.ModuleList()
        input_channels = 3
        for _ in range(num_layers):
            self.cnn_layers.append(nn.Conv2d(input_channels, 64, kernel_size=3, padding=1))
            self.cnn_layers.append(nn.BatchNorm2d(64))
            self.cnn_layers.append(nn.ReLU())
            self.cnn_layers.append(nn.MaxPool2d(kernel_size=2, stride=2))
            self.cnn_layers.append(nn.Dropout(dropout_rate))
            input_channels = 64

        self.feature_map_size = self._get_conv_output((3, 224, 224))
        self.fc1 = nn.Linear(self.feature_map_size + hog_feature_size, 128)
        self.fc2 = nn.Linear(128, num_classes)
        self.dropout = nn.Dropout(dropout_rate)

    def _get_conv_output(self, shape):
        x = torch.rand(1, *shape)
        x = self.forward_conv(x)
        return int(np.prod(x.size()))

    def forward_conv(self, x):
        for layer in self.cnn_layers:
            x = layer(x)
        return x

    def forward(self, x, hog_features):
        x = self.forward_conv(x)
        x = x.view(x.size(0), -1)
        x = torch.cat((x, hog_features), dim=1)
        x = torch.relu(self.fc1(x))
        x = self.dropout(x)
        x = self.fc2(x)
        return x

# Initialize the model, criterion, and optimizer
model = CNNModel(dropout_rate=0.3, num_layers=3, num_classes=3, hog_feature_size=hog_train.shape[1])
criterion = nn.CrossEntropyLoss(weight=class_weights_tensor)
optimizer = optim.Adam(model.parameters(), lr=0.001)
scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, 'min', factor=0.2, patience=5, verbose=True)

# Apply weights initialization
def weights_init(m):
    if isinstance(m, nn.Conv2d):
        nn.init.kaiming_normal_(m.weight.data, nonlinearity='relu')
    elif isinstance(m, nn.Linear):
        nn.init.xavier_normal_(m.weight.data)

model.apply(weights_init)

# Early stopping class
class EarlyStopping:
    def __init__(self, patience=10, delta=0):
        self.patience = patience
        self.delta = delta
        self.best_score = None
        self.early_stop = False
        self.counter = 0

    def __call__(self, val_loss, model, model_path):
        score = -val_loss
        if (self.best_score is None):
            self.best_score = score
            self.save_checkpoint(val_loss, model, model_path)
        elif (score < self.best_score + self.delta):
            self.counter += 1
            if (self.counter >= self.patience):
                self.early_stop = True
        else:
            self.best_score = score
            self.save_checkpoint(val_loss, model, model_path)
            self.counter = 0

    def save_checkpoint(self, val_loss, model, model_path):
        torch.save(model.state_dict(), model_path)

# Metrics class
class Metrics:
    def __init__(self):
        self.train_losses = []
        self.val_losses = []
        self.train_accuracies = []
        self.val_accuracies = []

    def append(self, train_loss, val_loss, train_acc, val_acc):
        self.train_losses.append(train_loss)
        self.val_losses.append(val_loss)
        self.train_accuracies.append(train_acc)
        self.val_accuracies.append(val_acc)

# Training function
def train_model(model, criterion, optimizer, train_loader, val_loader, num_epochs=50):
    model = model.to(device)
    best_val_loss = float('inf')
    train_losses = []
    val_losses = []
    train_accuracies = []
    val_accuracies = []
    
    early_stopping = EarlyStopping(patience=10, delta=0.01)

    for epoch in range(num_epochs):
        model.train()
        train_loss = 0.0
        train_corrects = 0
        
        for inputs, hog_features, labels in tqdm(train_loader):
            inputs = inputs.to(device, dtype=torch.float)
            hog_features = hog_features.to(device, dtype=torch.float)
            labels = labels.to(device, dtype=torch.long)
            
            optimizer.zero_grad()
            outputs = model(inputs, hog_features)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            
            train_loss += loss.item() * inputs.size(0)
            _, preds = torch.max(outputs, 1)
            train_corrects += torch.sum(preds == labels.data)
        
        epoch_loss = train_loss / len(train_loader.dataset)
        epoch_acc = train_corrects.double() / len(train_loader.dataset)
        
        train_losses.append(epoch_loss)
        train_accuracies.append(epoch_acc.item())
        
        model.eval()
        val_loss = 0.0
        val_corrects = 0

        with torch.no_grad():
            for inputs, hog_features, labels in val_loader:
                inputs = inputs.to(device, dtype=torch.float)
                hog_features = hog_features.to(device, dtype=torch.float)
                labels = labels.to(device, dtype=torch.long)
                
                outputs = model(inputs, hog_features)
                loss = criterion(outputs, labels)
                
                val_loss += loss.item() * inputs.size(0)
                _, preds = torch.max(outputs, 1)
                val_corrects += torch.sum(preds == labels.data)

        val_epoch_loss = val_loss / len(val_loader.dataset)
        val_epoch_acc = val_corrects.double() / len(val_loader.dataset)

        val_losses.append(val_epoch_loss)
        val_accuracies.append(val_epoch_acc.item())
        
        print(f'Epoch {epoch}/{num_epochs - 1}')
        print(f'Train Loss: {epoch_loss:.4f} Acc: {epoch_acc:.4f}')
        print(f'Val Loss: {val_epoch_loss:.4f} Acc: {val_epoch_acc:.4f}')

        scheduler.step(val_epoch_loss)
        early_stopping(val_epoch_loss, model, model_save_path)
        if early_stopping.early_stop:
            print("Early stopping")
            break

    metrics = Metrics()
    metrics.train_losses = train_losses
    metrics.val_losses = val_losses
    metrics.train_accuracies = train_accuracies
    metrics.val_accuracies = val_accuracies

    return metrics, model

# Device configuration
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Train the final model on the training data
final_metrics, final_model = train_model(model, criterion, optimizer, train_loader, val_loader, num_epochs=50)

# Evaluate the final model
def evaluate_model(model, test_loader):
    model = model.to(device)
    model.eval()
    test_loss = 0.0
    test_corrects = 0
    all_preds = []
    all_labels = []

    with torch.no_grad():
        for inputs, hog_features, labels in test_loader:
            inputs = inputs.to(device, dtype=torch.float)
            hog_features = hog_features.to(device, dtype=torch.float)
            labels = labels.to(device, dtype=torch.long)
            
            outputs = model(inputs, hog_features)
            loss = criterion(outputs, labels)
            
            test_loss += loss.item() * inputs.size(0)
            _, preds = torch.max(outputs, 1)
            test_corrects += torch.sum(preds == labels.data)
            
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    epoch_loss = test_loss / len(test_loader.dataset)
    epoch_acc = test_corrects.double() / len(test_loader.dataset)

    print(f'Test Loss: {epoch_loss:.4f} Acc: {epoch_acc:.4f}')
    
    return all_labels, all_preds

# Evaluate the trained model on the test set
y_true, y_pred = evaluate_model(final_model, test_loader)

# Convert numpy int64 to Python int for JSON serialization
y_true = [int(i) for i in y_true]
y_pred = [int(i) for i in y_pred]

# Calculate additional metrics
precision, recall, f1, _ = precision_recall_fscore_support(y_true, y_pred, average='weighted')
metrics = {
    'accuracy': accuracy_score(y_true, y_pred),
    'precision': precision,
    'recall': recall,
    'f1_score': f1,
    'confusion_matrix': confusion_matrix(y_true, y_pred).tolist()
}

with open(metrics_save_path, 'w') as f:
    json.dump(metrics, f, indent=4)

print(f"Accuracy: {metrics['accuracy']:.4f}")
print(f"Precision: {metrics['precision']:.4f}")
print(f"Recall: {metrics['recall']:.4f}")
print(f"F1 Score: {metrics['f1_score']:.4f}")

# Plot confusion matrix
conf_matrix = confusion_matrix(y_true, y_pred)
target_names = list(label_map.keys())

# Ensure that the number of classes matches the confusion matrix dimensions
if conf_matrix.shape[0] != len(target_names):
    raise ValueError("Mismatch between the size of the confusion matrix and the number of labels")

plt.figure(figsize=(10, 7))
disp = ConfusionMatrixDisplay(confusion_matrix=conf_matrix, display_labels=target_names)
disp.plot(cmap=plt.cm.Blues)
plt.title('Confusion Matrix')
plt.savefig("CNN_results/3c_side_CNN_confusion_matrix.png")
plt.show()

# Plot bar plot of actual vs. predicted labels
def plot_bar(y_true, y_pred, label_map, plot_save_path):
    unique_labels = list(label_map.values())
    correct_counts = [np.sum((np.array(y_true) == label) & (np.array(y_pred) == label)) for label in unique_labels]
    incorrect_counts = [np.sum((np.array(y_true) == label) & (np.array(y_pred) != label)) for label in unique_labels]

    bar_width = 0.35
    index = np.arange(len(unique_labels))

    plt.figure(figsize=(10, 6))
    plt.bar(index, correct_counts, bar_width, label='Correct', color='g')
    plt.bar(index + bar_width, incorrect_counts, bar_width, label='Incorrect', color='r')

    plt.xlabel('Labels')
    plt.ylabel('Counts')
    plt.title('Correct vs Incorrect Predictions')
    plt.xticks(index + bar_width / 2, list(label_map.keys()))
    plt.ylim(0, max(max(correct_counts), max(incorrect_counts)) + 10)  # Ensure y-axis starts at 0
    plt.legend()
    plt.tight_layout()
    plt.savefig(plot_save_path)
    plt.show()

plot_bar(y_true, y_pred, label_map, bar_plot_save_path)

# Save test results
test_results_dict = {
    'y_test': y_true,
    'y_pred_test': y_pred
}
with open(results_save_path, 'w') as f:
    json.dump(test_results_dict, f)

print("Training and evaluation complete. Results saved.")

# Randomize the labels and evaluate the model again
randomized_labels = np.random.permutation(y_test)

# Evaluate the model on the test set with randomized labels
y_true_randomized, y_pred_randomized = evaluate_model(final_model, test_loader)

# Calculate additional metrics for randomized labels
precision_rand, recall_rand, f1_rand, _ = precision_recall_fscore_support(randomized_labels, y_pred_randomized, average='weighted')
randomized_metrics = {
    'accuracy': accuracy_score(randomized_labels, y_pred_randomized),
    'precision': precision_rand,
    'recall': recall_rand,
    'f1_score': f1_rand,
    'confusion_matrix': confusion_matrix(randomized_labels, y_pred_randomized).tolist()
}

with open(randomized_metrics_save_path, 'w') as f:
    json.dump(randomized_metrics, f, indent=4)

print(f"Randomized Accuracy: {randomized_metrics['accuracy']:.4f}")
print(f"Randomized Precision: {randomized_metrics['precision']:.4f}")
print(f"Randomized Recall: {randomized_metrics['recall']:.4f}")
print(f"Randomized F1 Score: {randomized_metrics['f1_score']:.4f}")

# Plot bar plot of actual vs. predicted labels with randomized labels
plot_bar(randomized_labels, y_pred_randomized, label_map, randomized_bar_plot_save_path)

# table with metrics:
# Evaluate the trained model on the test set
y_true, y_pred = evaluate_model(final_model, test_loader)

# Convert numpy int64 to Python int for JSON serialization
y_true = [int(i) for i in y_true]
y_pred = [int(i) for i in y_pred]

# Calculate additional metrics
precision, recall, f1, support = precision_recall_fscore_support(y_true, y_pred, average=None)
accuracy = accuracy_score(y_true, y_pred)

# Create a DataFrame for the F1 scores by class
f1_score_table = pd.DataFrame({
    'Class': list(label_map.keys()),
    'Precision': precision,
    'Recall': recall,
    'F1 Score': f1,
    'Support': support
})

# Save the F1 score table to a CSV file
f1_score_table_path = 'CNN_results/3c_side_f1_score_table.csv'
f1_score_table.to_csv(f1_score_table_path, index=False)

print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision.mean():.4f}")
print(f"Recall: {recall.mean():.4f}")
print(f"F1 Score: {f1.mean():.4f}")

print(f1_score_table)

# Plot confusion matrix
conf_matrix = confusion_matrix(y_true, y_pred)
target_names = list(label_map.keys())

# Ensure that the number of classes matches the confusion matrix dimensions
if conf_matrix.shape[0] != len(target_names):
    raise ValueError("Mismatch between the size of the confusion matrix and the number of labels")

plt.figure(figsize=(10, 7))
disp = ConfusionMatrixDisplay(confusion_matrix=conf_matrix, display_labels=target_names)
disp.plot(cmap=plt.cm.Blues)
plt.title('Confusion Matrix')
plt.savefig("CNN_results/3c_side_CNN_confusion_matrix.png")
plt.show()

# Plot bar plot of actual vs. predicted labels
def plot_bar(y_true, y_pred, label_map, plot_save_path):
    unique_labels = list(label_map.values())
    correct_counts = [np.sum((np.array(y_true) == label) & (np.array(y_pred) == label)) for label in unique_labels]
    incorrect_counts = [np.sum((np.array(y_true) == label) & (np.array(y_pred) != label)) for label in unique_labels]

    bar_width = 0.35
    index = np.arange(len(unique_labels))

    plt.figure(figsize=(10, 6))
    plt.bar(index, correct_counts, bar_width, label='Correct', color='g')
    plt.bar(index + bar_width, incorrect_counts, bar_width, label='Incorrect', color='r')

    plt.xlabel('Labels')
    plt.ylabel('Counts')
    plt.title('Correct vs Incorrect Predictions')
    plt.xticks(index + bar_width / 2, list(label_map.keys()))
    plt.ylim(0, max(max(correct_counts), max(incorrect_counts)) + 10)  # Ensure y-axis starts at 0
    plt.legend()
    plt.tight_layout()
    plt.savefig(plot_save_path)
    plt.show()

plot_bar(y_true, y_pred, label_map, bar_plot_save_path)

# Save test results
test_results_dict = {
    'y_test': y_true,
    'y_pred_test': y_pred
}
with open(results_save_path, 'w') as f:
    json.dump(test_results_dict, f)

print("Training and evaluation complete. Results saved.")

# Randomize the labels and evaluate the model again
randomized_labels = np.random.permutation(y_test)

# Evaluate the model on the test set with randomized labels
y_true_randomized, y_pred_randomized = evaluate_model(final_model, test_loader)

# Calculate additional metrics for randomized labels
precision_rand, recall_rand, f1_rand, _ = precision_recall_fscore_support(randomized_labels, y_pred_randomized, average='weighted')
randomized_metrics = {
    'accuracy': accuracy_score(randomized_labels, y_pred_randomized),
    'precision': precision_rand,
    'recall': recall_rand,
    'f1_score': f1_rand,
    'confusion_matrix': confusion_matrix(randomized_labels, y_pred_randomized).tolist()
}

with open(randomized_metrics_save_path, 'w') as f:
    json.dump(randomized_metrics, f, indent=4)

print(f"Randomized Accuracy: {randomized_metrics['accuracy']:.4f}")
print(f"Randomized Precision: {randomized_metrics['precision']:.4f}")
print(f"Randomized Recall: {randomized_metrics['recall']:.4f}")
print(f"Randomized F1 Score: {randomized_metrics['f1_score']:.4f}")

# Plot bar plot of actual vs. predicted labels with randomized labels
plot_bar(randomized_labels, y_pred_randomized, label_map, randomized_bar_plot_save_path)
```
