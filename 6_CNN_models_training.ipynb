```python
import os
import numpy as np
import cv2
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
import matplotlib.pyplot as plt
from skimage.feature import hog
from tqdm import tqdm
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import concurrent.futures
import json

# Paths to the folders containing the images and height arrays
top_image_folder = 'brispi/top_images'
height_folder = 'height_arrays'
output_dir = 'demo_results/demo_CNN_k_segments'
model_save_path = os.path.join(output_dir, 'demo_k-segments_top_model.pth')
test_image_save_path = os.path.join(output_dir, 'k-segments_test_images.csv')

# Ensure the output directory exists
os.makedirs(output_dir, exist_ok=True)

# Device configuration
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Best hyperparameters
best_params = {
    'dropout_rate': 0.3080828194802809,
    'lr': 4.4624454489403664e-05,
    'batch_size': 64,
    'num_layers': 6,
    'num_segments': 3,
    'fc_units': 289
}

random_state = 42

# Common Functions
def split_filename(filename):
    parts = filename.split("_")
    base = "_".join(parts[:5])
    z = "_".join(parts[5:7])
    y_m_d_h = "_".join(parts[8:12])
    y_m_d, h = y_m_d_h.split("_")[0:3], y_m_d_h.split("_")[-1]
    min_sec = "_".join(parts[12:])
    return base, z, y_m_d, min_sec, h

def get_files_and_splits(directory):
    files = os.listdir(directory)
    return [split_filename(f) + (f,) for f in files]

def hours_within_threshold(hour1, hour2, threshold=1):
    hour1, hour2 = int(hour1), int(hour2)
    return (hour1 == hour2) or ((hour1 + 1) % 24 == hour2) or ((hour1 - 1) % 24 == hour2)

def calculate_means(height_array, num_parts=5):
    part_size = len(height_array) // num_parts
    means = [np.mean(height_array[i * part_size:(i + 1) * part_size]) for i in range(num_parts)]
    return means

def load_image_and_height(match, img_size=224, num_segments=224):
    top_image_name, height_array_name = match
    img_path = os.path.join(top_image_folder, top_image_name)
    height_path = os.path.join(height_folder, height_array_name)

    img = cv2.imread(img_path)
    if img is None:
        return None, None

    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = cv2.rotate(img, cv2.ROTATE_180)

    height, width = img.shape[:2]
    start_x = (width - 3280) // 2
    start_y = (height - 2464) // 2
    img = img[start_y:start_y + 2464, start_x:start_x + 3280]
    img = cv2.resize(img, (img_size, img_size))

    height_array = np.load(height_path)
    if height_array.size == 0:
        return None, None

    normalized_height_array = height_array / 2464.0
    segment_length = len(normalized_height_array) // num_segments
    avg_height_per_segment = np.array([
        np.mean(normalized_height_array[i * segment_length:(i + 1) * segment_length])
        for i in range(num_segments)
    ])

    return img, avg_height_per_segment

def preprocess_data(matches, img_size=224, num_segments=224):
    images, height_arrays, matched_filenames = [], [], []
    with concurrent.futures.ThreadPoolExecutor() as executor:
        results = list(executor.map(lambda match: load_image_and_height(match, img_size, num_segments), matches))

    for match, (img, height_array) in zip(matches, results):
        if img is not None and height_array is not None:
            images.append(img)
            height_arrays.append(height_array)
            matched_filenames.append(match)

    return np.array(images), np.array(height_arrays), matched_filenames

def extract_hog_features(image):
    gray_image = (image * 255).astype(np.uint8)
    gray_image = cv2.cvtColor(gray_image, cv2.COLOR_RGB2GRAY)
    fd, hog_image = hog(gray_image, orientations=8, pixels_per_cell=(16, 16),
                        cells_per_block=(1, 1), visualize=True)
    return fd, hog_image

def visualize_images(images, filenames, num_samples=5, title="Images"):
    plt.figure(figsize=(15, 5))
    for i in range(num_samples):
        plt.subplot(1, num_samples, i+1)
        img_path = os.path.join(top_image_folder, filenames[i][0])
        img = cv2.imread(img_path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        img = cv2.rotate(img, cv2.ROTATE_180)

        height, width = img.shape[:2]
        start_x = (width - 3280) // 2
        start_y = (height - 2464) // 2
        img = img[start_y:start_y + 2464, start_x:start_x + 3280]
        img = cv2.resize(img, (224, 224))

        plt.imshow(img)
        plt.axis('off')
    plt.suptitle(title)
    plt.tight_layout()
    plt.show()

# Define the CNN model
class CNNModel(nn.Module):
    def __init__(self, dropout_rate, num_layers, num_segments, hog_feature_size, fc_units):
        super(CNNModel, self).__init__()
        self.cnn_layers = nn.ModuleList()
        input_channels = 3
        for _ in range(num_layers):
            self.cnn_layers.append(nn.Conv2d(input_channels, 64, kernel_size=3, padding=1))
            self.cnn_layers.append(nn.BatchNorm2d(64))
            self.cnn_layers.append(nn.ReLU())
            self.cnn_layers.append(nn.MaxPool2d(kernel_size=2, stride=2))
            self.cnn_layers.append(nn.Dropout(dropout_rate))
            input_channels = 64

        self.feature_map_size = self._get_conv_output((3, 224, 224))
        self.fc1 = nn.Linear(self.feature_map_size + hog_feature_size, fc_units)
        self.fc2 = nn.Linear(fc_units, num_segments)
        self.dropout = nn.Dropout(dropout_rate)

    def _get_conv_output(self, shape):
        x = torch.rand(1, *shape)
        x = self.forward_conv(x)
        return int(np.prod(x.size()))

    def forward_conv(self, x):
        for layer in self.cnn_layers:
            x = layer(x)
        return x

    def forward(self, x, hog_features):
        x = self.forward_conv(x)
        x = x.view(x.size(0), -1)
        x = torch.cat((x, hog_features), dim=1)
        x = torch.relu(self.fc1(x))
        x = self.dropout(x)
        x = self.fc2(x)
        return x

def evaluate_model(model, test_loader, criterion):
    model.eval()
    test_loss = 0.0
    all_preds = []
    all_labels = []
    test_image_names = []

    with torch.no_grad():
        for inputs, hog_features, labels, image_names in test_loader:
            inputs = inputs.to(device, dtype=torch.float)
            hog_features = hog_features.to(device, dtype=torch.float)
            labels = labels.to(device, dtype=torch.float)

            outputs = model(inputs, hog_features)
            loss = criterion(outputs, labels)

            test_loss += loss.item() * inputs.size(0)
            all_preds.extend(outputs.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())
            test_image_names.extend(image_names)

    epoch_loss = test_loss / len(test_loader.dataset)
    mse = mean_squared_error(all_labels, all_preds)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(all_labels, all_preds)
    r2 = r2_score(all_labels, all_preds)

    return {
        'test_loss': epoch_loss,
        'mean_squared_error': mse,
        'root_mean_squared_error': rmse,
        'mean_absolute_error': mae,
        'r2_score': r2,
        'preds': np.array(all_preds),
        'labels': np.array(all_labels),
        'image_names': test_image_names
    }

class HeightDataset(Dataset):
    def __init__(self, images, hog_features, labels, image_names, transform=None):
        self.images = images
        self.hog_features = hog_features
        self.labels = labels
        self.image_names = image_names
        self.transform = transform

    def __len__(self):
        return len(self.images)

    def __getitem__(self, idx):
        img = self.images[idx]
        hog_feature = self.hog_features[idx]
        label = self.labels[idx]
        image_name = self.image_names[idx]
        if self.transform:
            img = self.transform(img)
        return img, hog_feature, label, image_name

def convert_to_serializable(obj):
    if isinstance(obj, np.ndarray):
        return obj.tolist()
    if isinstance(obj, np.generic):
        return obj.item()
    return obj

# Get files and splits
image_files = get_files_and_splits(top_image_folder)
array_files = get_files_and_splits(height_folder)

# Match files based on base, y_m_d and hour
matches = [(image[-1], array[-1]) for image in image_files for array in array_files
           if image[0] == array[0] and image[2] == array[2] and hours_within_threshold(image[4], array[4])]

# Load and preprocess data
num_segments = best_params['num_segments']
images, height_arrays, matched_filenames = preprocess_data(matches, img_size=224, num_segments=num_segments)
if len(images) == 0 or len(height_arrays) == 0:
    raise ValueError("No images or height arrays were loaded. Please check the data folder and files.")

# Visualize a sample image before training
sample_image = images[0]
sample_hog_image = extract_hog_features(sample_image)[1]
visualize_images([sample_image], [matched_filenames[0]], num_samples=1, title="Sample Image with HOG Features")

images = images / 255.0

train_transform = transforms.Compose([
    transforms.ToPILImage(),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(20),
    transforms.RandomResizedCrop(224, scale=(0.8, 1.0)),
    transforms.ToTensor()
])

val_test_transform = transforms.Compose([
    transforms.ToPILImage(),
    transforms.ToTensor()
])

kf = KFold(n_splits=5, shuffle=True, random_state=random_state)

test_image_names = []
fold_metrics = []
all_train_losses = []
all_val_losses = []

for fold, (train_val_idx, test_idx) in enumerate(kf.split(images)):
    train_val_images, test_images = images[train_val_idx], images[test_idx]
    train_val_hog_features, test_hog_features = height_arrays[train_val_idx], height_arrays[test_idx]
    train_val_segment_means, test_segment_means = height_arrays[train_val_idx], height_arrays[test_idx]

    # Save the names of the test images
    test_image_names.extend([(fold, matched_filenames[idx]) for idx in test_idx])

    train_size = int(0.8 * len(train_val_images))
    val_size = len(train_val_images) - train_size

    train_indices = list(range(train_size))
    val_indices = list(range(train_size, len(train_val_images)))

    train_images, val_images = train_val_images[train_indices], train_val_images[val_indices]
    train_hog_features, val_hog_features = train_val_hog_features[train_indices], train_val_hog_features[val_indices]
    train_segment_means, val_segment_means = train_val_segment_means[train_indices], train_val_segment_means[val_indices]

    train_dataset = HeightDataset(
        images=train_images, 
        hog_features=train_hog_features, 
        labels=train_segment_means, 
        image_names=[matched_filenames[idx] for idx in train_indices],
        transform=train_transform
    )

    val_dataset = HeightDataset(
        images=val_images, 
        hog_features=val_hog_features, 
        labels=val_segment_means, 
        image_names=[matched_filenames[idx] for idx in val_indices],
        transform=val_test_transform
    )

    test_dataset = HeightDataset(
        images=test_images, 
        hog_features=test_hog_features, 
        labels=test_segment_means, 
        image_names=[matched_filenames[idx] for idx in test_idx],
        transform=val_test_transform
    )

    train_loader = DataLoader(train_dataset, batch_size=best_params['batch_size'], shuffle=True, num_workers=4)
    val_loader = DataLoader(val_dataset, batch_size=best_params['batch_size'], shuffle=False, num_workers=4)
    test_loader = DataLoader(test_dataset, batch_size=best_params['batch_size'], shuffle=False, num_workers=4)

    hog_feature_size = train_hog_features.shape[1]

    model = CNNModel(dropout_rate=best_params['dropout_rate'], num_layers=best_params['num_layers'], num_segments=num_segments, hog_feature_size=hog_feature_size, fc_units=best_params['fc_units']).to(device)
    optimizer = optim.Adam(model.parameters(), lr=best_params['lr'])
    criterion = nn.MSELoss().to(device)
    scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, 'min', factor=0.2, patience=5)

    num_epochs = 50
    best_val_loss = float('inf')

    train_losses = []
    val_losses = []

    for epoch in range(num_epochs):
        model.train()
        train_loss = 0.0

        for inputs, hog_features, labels, _ in tqdm(train_loader):  # Adjusted loop to expect 4 values
            inputs = inputs.to(device, dtype=torch.float)
            hog_features = hog_features.to(device, dtype=torch.float)
            labels = labels.to(device, dtype=torch.float)

            optimizer.zero_grad()
            outputs = model(inputs, hog_features)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            train_loss += loss.item() * inputs.size(0)

        epoch_loss = train_loss / len(train_loader.dataset)
        train_losses.append(epoch_loss)
        print(f'Epoch {epoch+1}, Train Loss: {epoch_loss:.4f}')

        # Validation step
        model.eval()
        val_loss = 0.0

        with torch.no_grad():
            for inputs, hog_features, labels, _ in val_loader:  # Adjusted loop to expect 4 values
                inputs = inputs.to(device, dtype=torch.float)
                hog_features = hog_features.to(device, dtype=torch.float)
                labels = labels.to(device, dtype=torch.float)

                outputs = model(inputs, hog_features)
                loss = criterion(outputs, labels)
                val_loss += loss.item() * inputs.size(0)

        val_loss /= len(val_loader.dataset)
        val_losses.append(val_loss)
        print(f'Epoch {epoch+1}, Validation Loss: {val_loss:.4f}')

        scheduler.step(val_loss)

        if val_loss < best_val_loss:
            best_val_loss = val_loss
            torch.save(model.state_dict(), model_save_path)

    all_train_losses.append(train_losses)
    all_val_losses.append(val_losses)

    print(f'Finished fold {fold + 1}')

# Plotting train and validation losses for each fold
plt.figure(figsize=(10, 8))
colors = ['b', 'g', 'r', 'c', 'm']
for fold in range(5):
    plt.plot(all_train_losses[fold], color=colors[fold], label=f'Train Loss Fold {fold+1}')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.title('Train Loss for Each Fold')
plt.savefig(os.path.join(output_dir, 'k-split_train_loss_per_fold.png'))
plt.show()

plt.figure(figsize=(10, 8))
for fold in range(5):
    plt.plot(all_val_losses[fold], color=colors[fold], linestyle='--', label=f'Validation Loss Fold {fold+1}')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.title('Validation Loss for Each Fold')
plt.savefig(os.path.join(output_dir, 'k-split_val_loss_per_fold.png'))
plt.show()

# Save the test image names to a CSV file
test_image_df = pd.DataFrame(test_image_names, columns=['Fold', 'ImageName'])
test_image_df.to_csv(test_image_save_path, index=False)

# Load the best model for testing
model.load_state_dict(torch.load(model_save_path))
test_results = evaluate_model(model, test_loader, criterion)
print(f'Test Loss: {test_results["test_loss"]:.4f}, MSE: {test_results["mean_squared_error"]:.4f}, RMSE: {test_results["root_mean_squared_error"]:.4f}, MAE: {test_results["mean_absolute_error"]:.4f}, R²: {test_results["r2_score"]:.4f}')

# Save the test results
test_results_serializable = {k: convert_to_serializable(v) for k, v in test_results.items()}
with open(os.path.join(output_dir, 'k-split_top_RF_test_results.json'), 'w') as f:
    json.dump(test_results_serializable, f)

# Save predictions and actual labels to CSV
predictions_df = pd.DataFrame({
    'ImageName': test_results['image_names'],
    'Predicted1': test_results['preds'][:, 0],
    'Predicted2': test_results['preds'][:, 1],
    'Predicted3': test_results['preds'][:, 2],
    'Actual1': test_results['labels'][:, 0],
    'Actual2': test_results['labels'][:, 1],
    'Actual3': test_results['labels'][:, 2]
})
predictions_df.to_csv(test_image_save_path, index=False)

# Scatter plot of actual vs predicted labels
plt.scatter(test_results['labels'], test_results['preds'])
plt.xlabel('Actual Labels')
plt.ylabel('Predicted Labels')
plt.title('Actual vs Predicted Labels')
plt.savefig(os.path.join(output_dir, 'k-split_scatterplot_actual_vs_predicted.png'))
plt.show()
```

```python
# Import necessary libraries
import os
import numpy as np
import cv2
import pandas as pd
from sklearn.metrics import ConfusionMatrixDisplay, accuracy_score, confusion_matrix, precision_recall_fscore_support
from sklearn.model_selection import train_test_split
from sklearn.utils import class_weight
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
import matplotlib.pyplot as plt
from skimage import exposure
from skimage.feature import hog
from tqdm import tqdm
import concurrent.futures
import json

# Paths to the folders containing the images and height arrays
image_folders = {
    'side': 'brispi/side_images',
    'top': 'brispi/top_images'
}
height_folder = 'height_arrays'
model_save_paths = {
    'side': 'demo_results/demo_CNN_3class_side/demo_CNN_3class_side_model.pth',
    'top': 'demo_results/demo_CNN_3class_top/demo_CNN_3class_top_model.pth'
}
results_save_paths = {
    'side': 'demo_results/demo_CNN_3class_side/results.json',
    'top': 'demo_results/demo_CNN_3class_top/results.json'
}
metrics_save_paths = {
    'side': 'demo_results/demo_CNN_3class_side/metrics.json',
    'top': 'demo_results/demo_CNN_3class_top/metrics.json'
}
plot_save_paths = {
    'side': 'demo_results/demo_CNN_3class_side/confusion_matrix.png',
    'top': 'demo_results/demo_CNN_3class_top/confusion_matrix.png'
}
bar_plot_save_paths = {
    'side': 'demo_results/demo_CNN_3class_side/actual_vs_predicted_bar_plot.png',
    'top': 'demo_results/demo_CNN_3class_top/actual_vs_predicted_bar_plot.png'
}

# Ensure that the output directories exist
for path in model_save_paths.values():
    os.makedirs(os.path.dirname(path), exist_ok=True)

# Function to calculate the trimmed mean of a height array
def calculate_trimmed_mean(height_array, trim_bottom=0.08805012610937063, trim_top=0.2893423891188367):
    sorted_height_array = np.sort(height_array)
    lower_bound = int(trim_bottom * len(sorted_height_array))
    upper_bound = int((1 - trim_top) * len(sorted_height_array))
    
    if trim_bottom + trim_top >= 1.0:
        raise ValueError("The sum of trim_bottom and trim_top must be less than 1.0")
    
    trimmed_array = sorted_height_array[lower_bound:upper_bound]
    trimmed_mean = np.mean(trimmed_array)
    
    return trimmed_mean

# Helper functions
def split_filename(filename):
    parts = filename.split("_")
    base = "_".join(parts[:5])
    z = "_".join(parts[5:7])
    y_m_d_h = "_".join(parts[8:12])
    y_m_d, h = y_m_d_h.split("_")[0:3], y_m_d_h.split("_")[-1]
    min_sec = "_".join(parts[12:])
    return base, z, y_m_d, min_sec, h

def get_files_and_splits(directory):
    files = os.listdir(directory)
    return [split_filename(f) + (f,) for f in files]

def hours_within_threshold(hour1, hour2, threshold=1):
    hour1, hour2 = int(hour1), int(hour2)
    return (hour1 == hour2) or ((hour1 + 1) % 24 == hour2) or ((hour1 - 1) % 24 == hour2)

def load_image_and_height(match, image_type):
    if image_type == 'side':
        img_file = match + '.jpg'
        height_file = match + '.npy'
    else:
        top_image_name, height_array_name = match
        img_file = top_image_name
        height_file = height_array_name
    
    img_path = os.path.join(image_folders[image_type], img_file)
    height_path = os.path.join(height_folder, height_file)

    if not os.path.exists(img_path):
        print(f"Image file does not exist: {img_path}")
        return None, None, None

    img = cv2.imread(img_path)
    if img is None:
        print(f"Failed to load image: {img_path}")
        return None, None, None
    
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    if image_type == 'top':
        img = cv2.rotate(img, cv2.ROTATE_180)
    height, width = img.shape[:2]
    start_x = (width - 3280) // 2
    start_y = (height - 2464) // 2
    img = img[start_y:start_y + 2464, start_x:start_x + 3280]
    img = cv2.resize(img, (224, 224))

    height_array = np.load(height_path)
    if height_array.size == 0:
        print(f"Height array is empty: {height_path}")
        return None, None, None

    mean_height = calculate_trimmed_mean(height_array)
    hog_features, _ = extract_hog_features(img)
    return img, mean_height, hog_features

def preprocess_data(image_type):
    if image_type == 'side':
        matches = [f.replace('.jpg', '') for f in os.listdir(image_folders[image_type]) if f.endswith('.jpg')]
    else:
        image_files = get_files_and_splits(image_folders[image_type])
        array_files = get_files_and_splits(height_folder)
        matches = [(image[-1], array[-1]) for image in image_files for array in array_files
                   if image[0] == array[0] and image[2] == array[2] and hours_within_threshold(image[4], array[4])]
    
    images, mean_heights, hog_features_list = [], [], []
    with concurrent.futures.ThreadPoolExecutor() as executor:
        results = list(executor.map(lambda match: load_image_and_height(match, image_type), matches))
    for img, mean_height, hog_features in results:
        if img is not None and mean_height is not None:
            images.append(img)
            mean_heights.append(mean_height)
            hog_features_list.append(hog_features)
    return np.array(images), np.array(mean_heights), np.array(hog_features_list)

def custom_preprocessing(img):
    img = np.array(img)
    for channel in range(img.shape[2]):
        img[..., channel] = exposure.rescale_intensity(img[..., channel], out_range=(0, 1))
        img[..., channel] = exposure.equalize_hist(img[..., channel])
        img[..., channel] = exposure.equalize_adapthist(img[..., channel], clip_limit=0.03)
    return img

def extract_hog_features(img):
    hog_features, hog_image = hog(img, orientations=9, pixels_per_cell=(8, 8), cells_per_block=(2, 2), visualize=True, channel_axis=-1)
    return hog_features, hog_image

# Adjusted thresholds for better balance
adjusted_short_threshold = 748
adjusted_tall_threshold = 1205

# Label each height array using the adjusted thresholds
def label_height_array(mean_height, short_threshold, tall_threshold):
    if mean_height < short_threshold:
        return 'Short'
    elif mean_height > tall_threshold:
        return 'Tall'
    else:
        return 'Medium'

# PyTorch Dataset class
class HeightDataset(Dataset):
    def __init__(self, images, hog_features, labels, transform=None):
        self.images = images
        self.hog_features = hog_features
        self.labels = labels
        self.transform = transform

    def __len__(self):
        return len(self.images)

    def __getitem__(self, idx):
        img = self.images[idx]
        hog_feature = self.hog_features[idx]
        label = self.labels[idx]
        if self.transform:
            img = self.transform(img)
        return img, hog_feature, label

# Transforms with additional augmentations
train_transform = transforms.Compose([
    transforms.ToPILImage(),
    transforms.RandomHorizontalFlip(),
    transforms.Lambda(custom_preprocessing),
    transforms.ToTensor()
])

val_test_transform = transforms.Compose([
    transforms.ToPILImage(),
    transforms.Lambda(custom_preprocessing),
    transforms.ToTensor()
])

# Define the CNN model
class CNNModel(nn.Module):
    def __init__(self, dropout_rate=0.3, num_layers=3, num_classes=3, hog_feature_size=324):
        super(CNNModel, self).__init__()
        self.cnn_layers = nn.ModuleList()
        input_channels = 3
        for _ in range(num_layers):
            self.cnn_layers.append(nn.Conv2d(input_channels, 64, kernel_size=3, padding=1))
            self.cnn_layers.append(nn.BatchNorm2d(64))
            self.cnn_layers.append(nn.ReLU())
            self.cnn_layers.append(nn.MaxPool2d(kernel_size=2, stride=2))
            self.cnn_layers.append(nn.Dropout(dropout_rate))
            input_channels = 64

        self.feature_map_size = self._get_conv_output((3, 224, 224))
        self.fc1 = nn.Linear(self.feature_map_size + hog_feature_size, 128)
        self.fc2 = nn.Linear(128, num_classes)
        self.dropout = nn.Dropout(dropout_rate)

    def _get_conv_output(self, shape):
        x = torch.rand(1, *shape)
        x = self.forward_conv(x)
        return int(np.prod(x.size()))

    def forward_conv(self, x):
        for layer in self.cnn_layers:
            x = layer(x)
        return x

    def forward(self, x, hog_features):
        x = self.forward_conv(x)
        x = x.view(x.size(0), -1)
        x = torch.cat((x, hog_features), dim=1)
        x = torch.relu(self.fc1(x))
        x = self.dropout(x)
        x = self.fc2(x)
        return x

# Initialize the model, criterion, and optimizer
def initialize_model(hog_feature_size):
    model = CNNModel(dropout_rate=0.3, num_layers=3, num_classes=3, hog_feature_size=hog_feature_size)
    criterion = nn.CrossEntropyLoss(weight=class_weights_tensor)
    optimizer = optim.Adam(model.parameters(), lr=0.001)
    scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, 'min', factor=0.2, patience=5, verbose=True)

    def weights_init(m):
        if isinstance(m, nn.Conv2d):
            nn.init.kaiming_normal_(m.weight.data, nonlinearity='relu')
        elif isinstance(m, nn.Linear):
            nn.init.xavier_normal_(m.weight.data)

    model.apply(weights_init)
    return model, criterion, optimizer, scheduler

# Early stopping class
class EarlyStopping:
    def __init__(self, patience=10, delta=0):
        self.patience = patience
        self.delta = delta
        self.best_score = None
        self.early_stop = False
        self.counter = 0

    def __call__(self, val_loss, model, model_path):
        score = -val_loss
        if self.best_score is None:
            self.best_score = score
            self.save_checkpoint(val_loss, model, model_path)
        elif score < self.best_score + self.delta:
            self.counter += 1
            if self.counter >= self.patience:
                self.early_stop = True
        else:
            self.best_score = score
            self.save_checkpoint(val_loss, model, model_path)
            self.counter = 0

    def save_checkpoint(self, val_loss, model, model_path):
        torch.save(model.state_dict(), model_path)

# Metrics class
class Metrics:
    def __init__(self):
        self.train_losses = []
        self.val_losses = []
        self.train_accuracies = []
        self.val_accuracies = []

    def append(self, train_loss, val_loss, train_acc, val_acc):
        self.train_losses.append(train_loss)
        self.val_losses.append(val_loss)
        self.train_accuracies.append(train_acc)
        self.val_accuracies.append(val_acc)

# Training function
def train_model(model, criterion, optimizer, scheduler, train_loader, val_loader, num_epochs=50, model_save_path=''):
    model = model.to(device)
    best_val_loss = float('inf')
    train_losses = []
    val_losses = []
    train_accuracies = []
    val_accuracies = []

    early_stopping = EarlyStopping(patience=10, delta=0.01)

    for epoch in range(num_epochs):
        model.train()
        train_loss = 0.0
        train_corrects = 0

        for inputs, hog_features, labels in tqdm(train_loader):
            inputs = inputs.to(device, dtype=torch.float)
            hog_features = hog_features.to(device, dtype=torch.float)
            labels = labels.to(device, dtype=torch.long)

            optimizer.zero_grad()
            outputs = model(inputs, hog_features)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            train_loss += loss.item() * inputs.size(0)
            _, preds = torch.max(outputs, 1)
            train_corrects += torch.sum(preds == labels.data)

        epoch_loss = train_loss / len(train_loader.dataset)
        epoch_acc = train_corrects.double() / len(train_loader.dataset)

        train_losses.append(epoch_loss)
        train_accuracies.append(epoch_acc.item())

        model.eval()
        val_loss = 0.0
        val_corrects = 0

        with torch.no_grad():
            for inputs, hog_features, labels in val_loader:
                inputs = inputs.to(device, dtype=torch.float)
                hog_features = hog_features.to(device, dtype=torch.float)
                labels = labels.to(device, dtype=torch.long)

                outputs = model(inputs, hog_features)
                loss = criterion(outputs, labels)

                val_loss += loss.item() * inputs.size(0)
                _, preds = torch.max(outputs, 1)
                val_corrects += torch.sum(preds == labels.data)

        val_epoch_loss = val_loss / len(val_loader.dataset)
        val_epoch_acc = val_corrects.double() / len(val_loader.dataset)

        val_losses.append(val_epoch_loss)
        val_accuracies.append(val_epoch_acc.item())

        print(f'Epoch {epoch}/{num_epochs - 1}')
        print(f'Train Loss: {epoch_loss:.4f} Acc: {epoch_acc:.4f}')
        print(f'Val Loss: {val_epoch_loss:.4f} Acc: {val_epoch_acc:.4f}')

        scheduler.step(val_epoch_loss)
        early_stopping(val_epoch_loss, model, model_save_path)
        if early_stopping.early_stop:
            print("Early stopping")
            break

    metrics = Metrics()
    metrics.train_losses = train_losses
    metrics.val_losses = val_losses
    metrics.train_accuracies = train_accuracies
    metrics.val_accuracies = val_accuracies

    return metrics, model

# Device configuration
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Evaluate the final model
def evaluate_model(model, test_loader, criterion):
    model = model.to(device)
    model.eval()
    test_loss = 0.0
    test_corrects = 0
    all_preds = []
    all_labels = []

    with torch.no_grad():
        for inputs, hog_features, labels in test_loader:
            inputs = inputs.to(device, dtype=torch.float)
            hog_features = hog_features.to(device, dtype=torch.float)
            labels = labels.to(device, dtype=torch.long)
            
            outputs = model(inputs, hog_features)
            loss = criterion(outputs, labels)
            
            test_loss += loss.item() * inputs.size(0)
            _, preds = torch.max(outputs, 1)
            test_corrects += torch.sum(preds == labels.data)
            
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    epoch_loss = test_loss / len(test_loader.dataset)
    epoch_acc = test_corrects.double() / len(test_loader.dataset)

    print(f'Test Loss: {epoch_loss:.4f} Acc: {epoch_acc:.4f}')
    
    return all_labels, all_preds

# Plot bar plot of actual vs. predicted labels
def plot_bar(y_true, y_pred, label_map, plot_save_path):
    unique_labels = list(label_map.values())
    correct_counts = [np.sum((np.array(y_true) == label) & (np.array(y_pred) == label)) for label in unique_labels]
    incorrect_counts = [np.sum((np.array(y_true) == label) & (np.array(y_pred) != label)) for label in unique_labels]

    bar_width = 0.35
    index = np.arange(len(unique_labels))

    plt.figure(figsize=(10, 6))
    plt.bar(index, correct_counts, bar_width, label='Correct', color='g')
    plt.bar(index + bar_width, incorrect_counts, bar_width, label='Incorrect', color='r')

    plt.xlabel('Labels')
    plt.ylabel('Counts')
    plt.title('Correct vs Incorrect Predictions')
    plt.xticks(index + bar_width / 2, list(label_map.keys()))
    plt.ylim(0, max(max(correct_counts), max(incorrect_counts)) + 10)  # Ensure y-axis starts at 0
    plt.legend()
    plt.tight_layout()
    plt.savefig(plot_save_path)
    plt.show()

def run_cnn_demo(image_type='side'):
    # Preprocess data
    images, mean_heights, hog_features_list = preprocess_data(image_type)
    if len(images) == 0 or len(mean_heights) == 0:
        raise ValueError("No images or height arrays were loaded. Please check the data folder and files.")

    # Normalize images
    images = images / 255.0

    # Label the data
    labels = np.array([label_height_array(mean_height, adjusted_short_threshold, adjusted_tall_threshold) for mean_height in mean_heights])
    label_map = {'Short': 0, 'Medium': 1, 'Tall': 2}
    numerical_labels = np.array([label_map[label] for label in labels])

    # Check class balance
    unique, counts = np.unique(numerical_labels, return_counts=True)
    print(f"Class distribution ({image_type} data):", dict(zip(unique, counts)))

    # Ensure test set has one image from each class
    test_indices = []
    for class_label in np.unique(numerical_labels):
        class_indices = np.where(numerical_labels == class_label)[0]
        np.random.shuffle(class_indices)
        test_indices.append(class_indices[0])
    
    # The rest will be used for training
    train_val_indices = list(set(range(len(numerical_labels))) - set(test_indices))

    # Prepare the training and validation data
    X_train_val = images[train_val_indices]
    y_train_val = numerical_labels[train_val_indices]
    hog_train_val = hog_features_list[train_val_indices]

    X_test = images[test_indices]
    y_test = numerical_labels[test_indices]
    hog_test = hog_features_list[test_indices]

    X_train, X_val, y_train, y_val, hog_train, hog_val = train_test_split(X_train_val, y_train_val, hog_train_val, test_size=0.3, stratify=y_train_val, random_state=42)

    # Calculate class weights
    class_weights = class_weight.compute_class_weight(class_weight='balanced', classes=np.unique(y_train), y=y_train)
    global class_weights_tensor
    class_weights_tensor = torch.tensor(class_weights, dtype=torch.float)

    # Datasets and DataLoaders
    train_dataset = HeightDataset(X_train, hog_train, y_train, transform=train_transform)
    val_dataset = HeightDataset(X_val, hog_val, y_val, transform=val_test_transform)
    test_dataset = HeightDataset(X_test, hog_test, y_test, transform=val_test_transform)

    train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True, num_workers=4)
    val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False, num_workers=4)
    test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False, num_workers=4)

    # Initialize and train the model
    hog_feature_size = hog_train.shape[1]
    model, criterion, optimizer, scheduler = initialize_model(hog_feature_size)
    final_metrics, final_model = train_model(model, criterion, optimizer, scheduler, train_loader, val_loader, num_epochs=50, model_save_path=model_save_paths[image_type])

    # Evaluate the trained model on the test set
    y_true, y_pred = evaluate_model(final_model, test_loader, criterion)

    # Convert numpy int64 to Python int for JSON serialization
    y_true = [int(i) for i in y_true]
    y_pred = [int(i) for i in y_pred]

    # Calculate additional metrics
    precision, recall, f1, _ = precision_recall_fscore_support(y_true, y_pred, average='weighted')
    metrics = {
        'accuracy': accuracy_score(y_true, y_pred),
        'precision': precision,
        'recall': recall,
        'f1_score': f1,
        'confusion_matrix': confusion_matrix(y_true, y_pred).tolist()
    }

    with open(metrics_save_paths[image_type], 'w') as f:
        json.dump(metrics, f, indent=4)

    print(f"Accuracy: {metrics['accuracy']:.4f}")
    print(f"Precision: {metrics['precision']:.4f}")
    print(f"Recall: {metrics['recall']:.4f}")
    print(f"F1 Score: {metrics['f1_score']:.4f}")

    # Plot confusion matrix
    conf_matrix = confusion_matrix(y_true, y_pred)
    target_names = list(label_map.keys())

    # Ensure that the number of classes matches the confusion matrix dimensions
    if conf_matrix.shape[0] != len(target_names):
        raise ValueError("Mismatch between the size of the confusion matrix and the number of labels")

    plt.figure(figsize=(10, 7))
    disp = ConfusionMatrixDisplay(confusion_matrix=conf_matrix, display_labels=target_names)
    disp.plot(cmap=plt.cm.Blues)
    plt.title('Confusion Matrix')
    plt.savefig(plot_save_paths[image_type])
    plt.show()

    # Plot bar plot of actual vs. predicted labels
    plot_bar(y_true, y_pred, label_map, bar_plot_save_paths[image_type])

    # Save test results
    test_results_dict = {
        'y_test': y_true,
        'y_pred_test': y_pred
    }
    with open(results_save_paths[image_type], 'w') as f:
        json.dump(test_results_dict, f)

    print("Training and evaluation complete. Results saved.")

# Run the model for side images
run_cnn_demo('side')

# Run the model for top images
run_cnn_demo('top')
```
